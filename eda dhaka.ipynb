```python
import numpy as np
stock_data = df.copy()
count = 0

for col in stock_data:
    for i, data in enumerate(stock_data[col]):
        if data == "-":
            stock_data[col][i] = np.NaN
            count+=1
            print(f"{count}. Replaced with NaN in {i}th of {col}")
```
